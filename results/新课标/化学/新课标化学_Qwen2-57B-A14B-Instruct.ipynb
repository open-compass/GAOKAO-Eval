## Qwen2_57B-A14B 模型输出
反应①的反应类型为 **取代反应**；加入 $\text{K}_2\text{CO}_3$ 的作用是 **除去反应生成的酸** 和 **作为碱性催化剂**。

对照已知反应Ⅰ，反应③不使用 $\left(\text{C}_2\text{H}_5\right)_3\text{N}$ 也能进行，原因是 **反应Ⅰ中的 $\left(\text{C}_2\text{H}_5\right)_3\text{N}$ 可以被其他碱性物质替代**，如 $\text{K}_2\text{CO}_3$。

$E$ 中含氧官能团名称是 **酯基**。

$F$ 的结构简式是 **未给出**；反应⑤分两步进行，第一步产物的分子式为 $\text{C}_{15}\text{H}_{18}\text{F}_3\text{NO}_4$，其结构简式是 **未给出**。

$G$ 中手性碳原子是 **6**（写出序号）。

化合物 $H$ 是 $B$ 的同分异构体，具有苯环结构，核磁共振氢谱中显示为四组峰，且可以发生已知反应Ⅱ．则 $H$ 的可能结构是 **未给出**。


```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "Qwen/Qwen2-57B-A14B-Instruct"
gen_kwargs = {"max_length": 2048, "do_sample": False}

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True,
).eval()
```

```python
subject, paper_type = "化学", "新课标"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        text = tokenizer.apply_chat_template(
            [{"role": "user", "content": question}],
            tokenize=False,
            add_generation_prompt=True
        )
        model_inputs = tokenizer([text], return_tensors="pt").to(device)

        generated_ids = model.generate(model_inputs.input_ids, **gen_kwargs)
        generated_ids = [
            output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
        ]

        response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("Qwen2_57B-A14B 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
