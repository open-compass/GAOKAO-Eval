$  \overrightarrow { b } - 4 \overrightarrow { a } = ( 2 , x - 4 )$ ，$  \overrightarrow { b } \bot \left( \overrightarrow { b } - 4 \overrightarrow { a } \right)=0$ ， ∴$ \overrightarrow { b } \left( \overrightarrow { b } - 4 \overrightarrow { a } \right) = 0$ ， ∴$ 4+x(x-4)=0$ ， ∴$ x=2$ ． 故选$ \text{D}$ ．
## Qwen2 72B 模型输出
由题意知，$\overrightarrow{b} = (2, x)$，而$\overrightarrow{a} = (0, 1)$。根据题目条件，有$\overrightarrow{b} \bot (\overrightarrow{b} - 4\overrightarrow{a})$。

首先计算$\overrightarrow{b} - 4\overrightarrow{a}$：

\[
\overrightarrow{b} - 4\overrightarrow{a} = (2, x) - 4(0, 1) = (2, x) - (0, 4) = (2, x - 4)
\]

由于$\overrightarrow{b} \bot (\overrightarrow{b} - 4\overrightarrow{a})$，即$\overrightarrow{b}$与$(\overrightarrow{b} - 4\overrightarrow{a})$垂直，则它们的点积为0：

\[
\overrightarrow{b} \cdot (\overrightarrow{b} - 4\overrightarrow{a}) = 0
\]

将具体值代入上式：

\[
(2, x) \cdot (2, x - 4) = 0
\]

\[
2*2 + x*(x - 4) = 0
\]

\[
4 + x^2 - 4x = 0
\]

整理得：

\[
x^2 - 4x + 4 = 0
\]

这是一个标准的一元二次方程，可以写成完全平方的形式：

\[
(x - 2)^2 = 0
\]

解得：

\[
x = 2
\]

因此，正确答案是D. $2$。

已知：
1. $\cos(\alpha + \beta) = m$
2. $\tan\alpha\tan\beta = 2$

目标：求$\cos(\alpha - \beta)$

我们知道：
\[ \cos(\alpha + \beta) = \cos\alpha\cos\beta - \sin\alpha\sin\beta = m \]
\[ \cos(\alpha - \beta) = \cos\alpha\cos\beta + \sin\alpha\sin\beta \]

由$\tan\alpha\tan\beta = 2$，可以得到：
\[ \frac{\sin\alpha\sin\beta}{\cos\alpha\cos\beta} = 2 \]
\[ \Rightarrow \sin\alpha\sin\beta = 2\cos\alpha\cos\beta \]

将$\sin\alpha\sin\beta = 2\cos\alpha\cos\beta$代入$\cos(\alpha - \beta)$的表达式中，得到：
\[ \cos(\alpha - \beta) = \cos\alpha\cos\beta + 2\cos\alpha\cos\beta = 3\cos\alpha\cos\beta \]

\[ m = \cos\alpha\cos\beta - \sin\alpha\sin\beta \]
\[ m = \cos\alpha\cos\beta - 2\cos\alpha\cos\beta \]
\[ m = -\cos\alpha\cos\beta \]

但是这里有一个小错误，正确的推导应该是：
\[ m = \cos\alpha\cos\beta - 2\cos\alpha\cos\beta \]

\[ m = \cos\alpha\cos\beta - 2\cos\alpha\cos\beta = -\cos\alpha\cos\beta \]
\[ \Rightarrow \cos\alpha\cos\beta = -m \]

因此，
\[ \cos(\alpha - \beta) = 3(-m) = -3m \]
