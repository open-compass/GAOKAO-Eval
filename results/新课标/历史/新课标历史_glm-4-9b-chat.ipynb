## 题目编号：1
## 题目标答
根据材料“大型墓葬规模大、随葬品丰富，出土了成套的石制、 陶制礼器等”可提取出随葬品种类多样且出现了礼器，结合所学可知，礼器出现于仰韶文化时期，而仰韶文化属于新石器时代，同时根据材料“大型墓葬规模大、随葬品丰富”可知出现了贫富分化，在新石器时代晚期，出现了贫富分化的现象，因此可知该遗址处于新石器时代晚期，D项正确； 根据材料可知，该遗址出现了随葬品和礼器，这说明当时已经出现了贫富分化，而旧石器时代，尚未出现贫富分化，因此排除 A，B 项； 根据材料“大型墓葬规模大、随葬品丰富，出土了成套的石制、陶制礼器等”可知，该遗址已经出现了陶器，而新石器时代早期，出现了陶器，因此不能确定该遗址一定处于新石器时代早期，排除 C 项。
答案是：D
## GLM4 9B 模型输出

根据题目描述，该史前遗址发掘出1300余座墓葬，其中大型墓葬规模大、随葬品丰富，出土了成套的石制、陶制礼器等。我们可以通过以下步骤来推断该遗址所处的时代：

1. **墓葬数量和类型**：墓葬数量较多，且存在大、中、小三类墓葬，这表明该遗址已经进入了有明确墓葬制度的阶段。

2. **随葬品**：大型墓葬随葬品丰富，包括石制、陶制礼器，这表明当时社会已经出现了较为复杂的社会结构和等级制度。

3. **排除选项**：
   - A. 旧石器时代早期：这个时期的人类主要以狩猎采集为主，社会结构简单，墓葬制度不发达，随葬品也较为简单。
   - B. 旧石器时代晚期：虽然这个时期的人类开始使用陶器，但社会结构仍然较为简单，墓葬制度也不如新石器时代发达。
   - C. 新石器时代早期：这个时期的人类开始出现农业，社会结构逐渐复杂，墓葬制度开始发展，但随葬品通常不如新石器时代晚期丰富。
   - D. 新石器时代晚期：这个时期的人类社会结构复杂，出现了明显的等级制度，墓葬制度发达，随葬品丰富。

综合以上分析，可以推断该遗址处于新石器时代晚期。

答案：D. 新石器时代晚期

## 题目标答
春秋战国时期，各国纷纷进行政治变革，争相延揽人才，诸子百家大都呼吁“选贤与能”，属于变法派，符合当时的政治需求，是历史发展趋势。儒家代表孔子和孟子就积极奔走各国宣传仁义之道；墨家代表墨子则倡导兼爱非攻，这些都有利于政治稳定，推动社会发展。但道家主张“无为而治”，主张“小国寡民”，不符合当时的社会现实，故 C符合题意。
## GLM4 9B 模型输出

在春秋战国时期，儒家、墨家、道家和法家都是诸子百家中重要的学派。其中，儒家、墨家和法家都主张“选贤与能”，即选拔有才能的人来治理国家。

A. 儒家：儒家主张仁政，认为君主应该以德治国，选拔贤能之人来辅佐自己，实现国家的和谐与繁荣。

B. 墨家：墨家主张兼爱、非攻，认为国家应该选拔有才能的人来治理，以实现社会的公平与正义。

D. 法家：法家主张法治，认为国家的治理应该依靠法律和制度，选拔有才能的人来执行法律，维护国家的稳定。

C. 道家：道家主张无为而治，认为君主应该顺应自然，减少干预，让人民自行发展。道家并不强调选拔贤能，而是强调顺应自然和道的原则。

因此，不认同“选贤与能”理念的学派是C. 道家。

```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "THUDM/glm-4-9b-chat"
gen_kwargs = {"max_length": 2048, "do_sample": False}

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    torch_dtype=torch.float16,
).eval().to(device)
```

```python
subject, paper_type = "历史", "新课标"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        inputs = tokenizer.apply_chat_template(
            [{"role": "user", "content": question}],
            add_generation_prompt=True,
            tokenize=True,
            return_tensors="pt",
            return_dict=True
        )
        inputs = inputs.to(device)

        with torch.no_grad():
            outputs = model.generate(**inputs, **gen_kwargs)
            outputs = outputs[:, inputs['input_ids'].shape[1]:]
            
            response = tokenizer.decode(outputs[0], skip_special_tokens=True)

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("GLM4 9B 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
