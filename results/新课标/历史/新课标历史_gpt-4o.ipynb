```python
import re
import json
import requests
model_path = "gpt-4o"
gen_kwargs = {"max_length": 2048, "do_sample": False}
OPENAI_API_KEY = ""
OPENAI_API_BASE_URL = "https://api.openai.com"
url = "https://api.openai.com/v1/chat/completions"
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {OPENAI_API_KEY}"
}



```

```python
subject, paper_type = "历史", "新课标"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        # inputs = tokenizer(question, return_tensors="pt")
        payload = {
            "model": model_path,
            "messages": [
                {
                    "role": "user",
                    "content": question
                }
            ],
            "max_tokens": 4096,
            "temperature": 0.0,
        }
        
        response = requests.post(url, headers=headers, json=payload)
        res_content = response.json()["choices"][0]["message"]["content"]
        response = res_content

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("gpt-4o 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": data['prompt'],
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
        import time

        # Sleep for 30 seconds
        time.sleep(30)
```
