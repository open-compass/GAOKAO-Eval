## 题目编号：1
## 题目标答
①：空想社会主义没有提出缔结反映各族人民意愿的国家联盟的主张，①排除。②④：新和谐公社是英国空想社会主义者欧文建立的理想模型，“和谐制度”构想是由法国空想社会主义者傅立叶提出，倡导建立人人劳动、普遍协作的“和谐制度”。虽然最终失败了，但也是一次有意义的尝试。这种尝试及其思想，为后来马克思主义的产生起到了重要作用，②④符合题意。③：空想社会主义是一种理论，其构想和原则并没有在实践中得到真正的实现；而巴黎公社则是一场实际的革命运动，它试图通过建立无产阶级政权来实现社会主义的理想。巴黎公社不是空想社会主义的主张也不是其实践活动，③排除。故本题选C。
## Yi 34B 模型输出
C
空想社会主义是现代社会主义思想的来源之一，空想共产主义都主张废除私有制，消灭阶级差别，共同劳动，平均分配产品，建立社会平等。他们中有些代表人物还主张暴力革命，并提出和论证了过渡时期等问题。但他们在一些基本问题上，还未能摆脱空想社会主义的根本缺陷，因而他们的理想社会不能实现。
①错误，“缔结反映各族人民意愿的国家联盟”不是空想社会主义的主张；
②④正确，进行共产主义的“新和谐公社”新村实验和倡导建立人人劳动、普遍协作的“和谐制度”是空想社会主义者为建立美好社会的努力；
③错误，成立被誉为新社会的光辉先驱的巴黎公社不是空想社会主义者的行为，是科学社会主义。
故选：C。
本题考查：空想社会主义
（1）积极性：空想社会主义者揭露资本主义的弊端，设计未来社会的美好蓝图。
（2）局限性：空想社会主义者仅从理性、正义等原则出发，主张阶级调和，反对阶级斗争，看不到广大人民群众特别是无产阶级的力量，没有找到消灭资本主义社会和建立新社会的强大力量，也没有找到进行社会变革的正确途径。
本题考查学生调动和运用基础知识解决问题的能力，目的是让学生通过对材料的分析来反思所学的内容，以此来加深知识的掌握程度，只要掌握好基础知识就可做好本题。

## 题目编号：12
## 题目标答
如果在地铁客流高峰时段提高票价，就能减轻客流压力。属于充分条件假言判断，其推理规则是有前必有后，无后必无前。A：题干只能表述出地铁客流高峰期时段提高票价是减轻客流压力的充分条件，不能判断出是不是必要条件，因此该判断不一定为假，A排除。B：根据充分条件假言判断有前必有后，则在地铁客流高峰期提高票价，一定能减轻客流压力，因此不能减轻客流压力为假，B符合题意。C：不在地铁客流高峰时段提高票价，前件假，后件不一定，因此也能减轻客流压力不一定为假，C排除。D：如果要减轻客流压力，即后件为真，前件不一定为假，即在地铁客流高峰时段提高票价不一定为假，D排除。故本题选B。
## Yi 34B 模型输出
根据题意，有人认为：如果在地铁客流高峰时段提高票价，就能减轻客流压力。这一观点可以表示为：如果A（地铁客流高峰时段提高票价），那么B（减轻客流压力）。

A选项表示：只有A，才B，即只有在地铁客流高峰时段提高票价，才能减轻客流压力。这与题干中的观点是一致的，所以A选项不必然为假。

B选项表示：虽然A，但是非B，即虽然在地铁客流高峰时段提高票价，但是不能减轻客流压力。这与题干中的观点是矛盾的，如果题干观点成立，那么B选项必定为假。

乙的请求是有法律依据的。根据《物权法》的相关规定，物权人对自己的不动产或者动产，依法享有占有、使用、收益和处分的权利。甲在其宅基地范围内翻建房屋是他的权利，但这种权利的行使不得危及他人物权的安全。在本案中，甲在翻建房屋时没有预留出合理的滴水空间，导致雨水沿房檐滴到乙家房屋外墙，影响了乙对其房屋的正常使用，甚至造成了乙房屋的损害。因此，乙请求甲采取措施防止雨水沿房檐滴到自己家的房屋外墙是有法律依据的。

综上所述，乙的请求是合理的，甲的辩称是不成立的。甲应当采取措施防止雨水沿房檐滴到乙家房屋外墙，并赔偿乙因此而遭受的损失。这不仅是对乙合法权益的保护，也是甲应当承担的法律责任。在行使个人权利的同时，我们应当尊重他人的合法权益，遵循社会公德，共同维护良好的社会秩序。


```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "01-ai/Yi-1.5-34B-Chat"
gen_kwargs = {"max_length": 2048, "do_sample": False}

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    torch_dtype=torch.float16,
).eval().to(device)
```

```python
subject, paper_type = "政治", "新课标"
file_name = f"../data/{paper_type}/{subject}.jsonl"

terminators = [
    tokenizer.eos_token_id,
    tokenizer.convert_tokens_to_ids("<|eot_id|>"),
    tokenizer.convert_tokens_to_ids("<|im_end|>")
]

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        input_ids = tokenizer.apply_chat_template(conversation=[{"role": "user", "content": question}], 
                                                tokenize=True, 
                                                add_generation_prompt=True, 
                                                return_tensors='pt')
        output_ids = model.generate(input_ids.to(device), eos_token_id=terminators, **gen_kwargs)
        response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("Yi 34B 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
