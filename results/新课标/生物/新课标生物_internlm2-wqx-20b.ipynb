## InternLM-WQX 模型输出
限制酶切割的化学键是磷酸二酯键。为保证 $N$ 基因能在菌株 $B2$ 中表达，在构建片段甲时，应将 $M1$ 与 $M2$ 片段分别插入质粒的Ⅰ和 Ⅱ、Ⅲ和Ⅳ酶切位点之间，原因是 $M1$ 与 $M2$ 片段分别含有 $N$ 基因的启动子和终止子。 $CRISPR/Cas9$ 技术可以切割细菌 $B1$ 基因组中与向导 $RNA$ 结合的 $DNA$。向导 $RNA$ 与 $B1$ 基因组 $DNA$ 互补配对可以形成的碱基对有 $G—C$ 和 $A—U$（或 $T—A$）。 用引物 $P1$ 和 $P2$ 进行 $PCR$ 可验证片段甲插入了细菌 $B1$ 基因组，所用的模板是细菌 $B2$ 的基因组；若用该模板与引物 $P3$ 和 $P4$ 进行 $PCR$，实验结果是扩增出 $N$ 基因片段。 与秸秆焚烧相比，利用高效降解纤维素的细菌处理秸秆的优点是减少环境污染、提高资源利用率。
答案是：磷酸二酯键; $M1$ 与 $M2$ 片段分别含有 $N$ 基因的启动子和终止子; $A—U$（或 $T—A$）; 细菌 $B2$ 的基因组; 扩增出 $N$ 基因片段; 减少环境污染、提高资源利用率


```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "internlm/internlm2-wqx-20b"
gen_kwargs = {"max_length": 2048, "do_sample": False}

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    torch_dtype=torch.float16,
).eval().to(device)
```

```python
subject, paper_type = "生物", "新课标"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        inputs = tokenizer(question, return_tensors="pt")
        input_ids = inputs["input_ids"].to(device)

        outputs = model.generate(input_ids, **gen_kwargs)
        outputs = outputs[0].cpu().tolist()[len(inputs["input_ids"][0]) :]

        response = tokenizer.decode(outputs, skip_special_tokens=True)
        response = response.split("<|im_end|>")[0]

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("InternLM-WQX 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
