### 当 $a = -2$ 时，求 $f(x)$ 的极值

$f'(x) = \frac{d}{dx}[(1 + 2x) \ln(1 + x) - x]$

使用乘积法则和链式法则，我们有：

$f'(x) = 2\ln(1 + x) + (1 + 2x) \cdot \frac{1}{1 + x} - 1$

化简得：

$f'(x) = 2\ln(1 + x) + \frac{1 + 2x}{1 + x} - 1$

$= 2\ln(1 + x) + \frac{1 + 2x - (1 + x)}{1 + x}$

$= 2\ln(1 + x) + \frac{x}{1 + x}$

令 $f'(x) = 0$，解方程：

$2\ln(1 + x) + \frac{x}{1 + x} = 0$

$2\ln(1 + x) = -\frac{x}{1 + x}$

由于 $\ln(1 + x)$ 是增函数，且在 $x > -1$ 时定义，我们可以两边同时取自然对数的指数来解这个方程：

$e^{2\ln(1 + x)} = e^{-\frac{x}{1 + x}}$

### 第一个不等式的证明

给定 $a + b \geqslant 3$，我们需要证明 $2a^2 + 2b^2 > a + b$。我们可以尝试通过调整不等式的形式来简化证明过程。

首先，将不等式两边同时乘以2，得到：
$4a^2 + 4b^2 > 2a + 2b$

然后，将左边的项重新组织为：
$2a^2 + 2b^2 + 2(a^2 + b^2) > 2a + 2b$

注意到 $a^2 + b^2 \geqslant 2ab$（这是由算术平均数-几何平均数不等式得出的），因此我们可以进一步推导：
$2a^2 + 2b^2 + 2(2ab) > 2a + 2b$
$2a^2 + 2b^2 + 4ab > 2a + 2b$

接下来，我们可以通过添加和减去相同的项来重组上述不等式，使其更易于比较：
$2a^2 - 2a + 2b^2 - 2b + 4ab > 0$
$2(a^2 - a + \frac{1}{4}) + 2(b^2 - b + \frac{1}{4}) + 3 > 0$

注意到 $a^2 - a + \frac{1}{4} = (a - \frac{1}{2})^2 \geqslant 0$ 和 $b^2 - b + \frac{1}{4} = (b - \frac{1}{2})^2 \geqslant 0$，因此：
$2(a - \frac{1}{2})^2 + 2(b - \frac{1}{2})^2 + 3 > 0$

由于 $(a - \frac{1}{2})^2$ 和 $(b - \frac{1}{2})^2$ 都是非负的，所以 $2(a - \frac{1}{2})^2 + 2(b - \frac{1}{2})^2 + 3 > 0$ 对所有实数 $a$ 和 $b$ 都成立。这证明了原不等式 $2a^2 + 2b^2 > a + b$ 成立。

### 第二个不等式的证明

对于 $|a - 2b^2| + |b - 2a^2| \geqslant 6$ 的证明，我们同样可以采用直接的方法。

首先，考虑绝对值的性质，我们知道 $|x| \geqslant x$ 对于所有实数 $x$ 成立。因此，我们可以分别对 $a - 2b^2$ 和 $b - 2a^2$ 应用这个性质：

$|a - 2b^2| \geqslant a - 2b^2$
$|b - 2a^2| \geqslant b - 2a^2$

将这两个不等式相加，得到：
$|a - 2b^2| + |b - 2a^2| \geqslant (a - 2b^2) + (b - 2a^2)$
$= a + b - 2(a^2 + b^2)$

根据题目条件 $a + b \geqslant 3$，我们需要证明的是 $a + b - 2(a^2 + b^2) \geqslant 6$。这可以重写为：
$a + b \geqslant 2(a^2 + b^2) + 3$

考虑到 $a^2 + b^2 \geqslant \frac{(a + b)^2}{2}$（这是由算术平均数-几何平均数不等式得出的），我们可以继续推导：
$a + b \geqslant 2(\frac{(a + b)^2}{2}) + 3$
$a + b \geqslant (a + b)^2 + 3$

但是，这显然不是真的，因为我们知道 $(a + b)^2 \geqslant 9$（因为 $a + b \geqslant 3$），所以 $(a + b)^2 + 3 > a + b$。因此，我们需要寻找另一种方法来证明第二个不等式。

考虑到 $|a - 2b^2|$ 和 $|b - 2a^2|$ 可能取正值也可能取负值，我们可以构造一个最坏情况的场景来证明这个不等式。最坏的情况是当 $a - 2b^2$ 和 $b - 2a^2$ 同时为负数时，即 $a < 2b^2$ 和 $b < 2a^2$。在这种情况下，我们有：
$|a - 2b^2| + |b - 2a^2| = -(a - 2b^2) - (b - 2a^2)$
$= -a + 2b^2 - b + 2a^2$
$= 2a^2 - a + 2b^2 - b$

我们再次应用算术平均数-几何平均数不等式，得到：
$2a^2 - a + 2b^2 - b \geqslant 2\sqrt{2a^2 \cdot 2b^2} - a - b$
$= 4ab - a - b$

由于 $a + b \geqslant 3$，我们设 $a + b = k$，其中 $k \geqslant 3$。那么 $4ab \leqslant 2(a^2 + b^2)$，因此：
$4ab - a - b \leqslant 2(a^2 + b^2) - a - b$

我们需要证明的是 $2(a^2 + b^2) - a - b \geqslant 6$。由于 $a + b \geqslant 3$，我们可以将 $a^2 + b^2$ 表达为 $(a + b)^2 - 2ab$，从而：
$2((a + b)^2 - 2ab) - a - b$
$= 2(k^2 - 2ab) - k$
$= 2k^2 - 4ab - k$

由于 $4ab \leqslant 2k$（因为 $a + b = k$），我们可以进一步简化为：
$2k^2 - 4ab - k \geqslant 2k^2 - 2k - k$
$= 2k^2 - 3k$

由于 $k \geqslant 3$，我们可以看到 $2k^2 - 3k$ 是一个关于 $k$ 的二次函数，其在 $k \geqslant 3$ 上是递增的，并且在 $k = 3$ 时取得最小值 $2(3)^2 - 3(3) = 18 - 9 = 9 > 6$。因此，对于所有 $k \geqslant 3$，都有 $2k^2 - 3k \geqslant 6$，这就证明了 $|a - 2b^2| + |b - 2a^2| \geqslant 6$ 成立。


```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "Qwen/Qwen2-57B-A14B-Instruct"
gen_kwargs = {"max_length": 2048, "do_sample": False}

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    torch_dtype=torch.float16
    device_map="auto",
    trust_remote_code=True,
).eval()
```

```python
subject, paper_type = "理综数学", "全国甲卷"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        text = tokenizer.apply_chat_template(
            [{"role": "user", "content": question}],
            tokenize=False,
            add_generation_prompt=True
        )
        model_inputs = tokenizer([text], return_tensors="pt").to(device)

        generated_ids = model.generate(model_inputs.input_ids, **gen_kwargs)
        generated_ids = [
            output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
        ]

        response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("Qwen2_57B-A14B 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
