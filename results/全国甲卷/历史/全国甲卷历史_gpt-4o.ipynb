   - **经济全球化**：20世纪见证了国际贸易和投资的迅猛增长。布雷顿森林体系的建立、关税及贸易总协定（GATT）的签订以及后来的世界贸易组织（WTO）的成立，推动了全球经济一体化。跨国公司和全球供应链的兴起，使得各国经济更加紧密地联系在一起。
   - **信息全球化**：随着互联网的普及和信息技术的进步，信息传播的速度和广度大大增加。人们可以即时获取全球各地的信息，促进了文化交流和思想传播。

2. **科技迅猛发展**：
   - **工业技术进步**：20世纪初的第二次工业革命和中期的第三次工业革命带来了电力、汽车、飞机、计算机等一系列重大技术突破，极大地改变了人类的生产和生活方式。
   - **信息技术革命**：20世纪下半叶，计算机技术和互联网的迅速发展，标志着信息时代的到来。信息技术不仅改变了经济结构，也深刻影响了社会生活和文化形态。

3. **政治格局剧变**：
   - **两次世界大战**：20世纪上半叶的两次世界大战深刻改变了世界政治格局。第一次世界大战导致了欧洲旧帝国的崩溃，第二次世界大战则催生了美苏两极格局。
   - **冷战与解体**：二战后，美苏两大超级大国的对立形成了冷战格局。冷战期间，世界被分为两大阵营，直到1991年苏联解体，冷战结束，世界进入了多极化时代。
   - **殖民地独立运动**：二战后，亚洲、非洲和拉丁美洲的殖民地纷纷争取独立，形成了大批新兴国家，改变了全球政治版图。

4. **社会文化多样化**：
   - **社会变革**：20世纪见证了许多社会运动，如民权运动、女权运动、环保运动等，这些运动推动了社会的进步和变革，促进了社会的多样性和包容性。
   - **文化交流**：全球化和信息技术的发展促进了不同文化之间的交流与融合。西方文化的传播和本土文化的复兴，使得世界文化呈现出多样化的趋势。

综上所述，1900年到2000年，世界在经济、科技、政治和社会文化等方面都发生了深刻的变化。这些变化不仅塑造了现代世界的面貌，也为21世纪的发展奠定了基础。

```python
subject, paper_type = "历史", "全国甲卷"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        # inputs = tokenizer(question, return_tensors="pt")
        payload = {
            "model": model_path,
            "messages": [
                {
                    "role": "user",
                    "content": question
                }
            ],
            "max_tokens": 4096,
            "temperature": 0.0,
        }
        
        response = requests.post(url, headers=headers, json=payload)
        res_content = response.json()["choices"][0]["message"]["content"]
        response = res_content

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("gpt-4o 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": data['prompt'],
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
        import time

        # Sleep for 30 seconds
        time.sleep(30)
```
