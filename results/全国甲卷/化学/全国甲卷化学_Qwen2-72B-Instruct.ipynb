### 放电过程

在放电过程中，锌（$\mathrm{Zn}$）作为负极发生氧化反应，而二氧化锰（$\mathrm{MnO}_{2}$）作为正极发生还原反应。具体反应如下：

这与选项C描述的正极反应一致，因此C是正确的。

在充电过程中，上述反应逆向进行。这意味着：

- 阳极（原正极，现在发生氧化反应）：$\text{MnOOH} + \mathrm{OH}^{-} \rightarrow \mathrm{MnO}_{2} + \mathrm{H}_{2}\mathrm{O} + \mathrm{e}^{-}$
- 阴极（原负极，现在发生还原反应）：$\mathrm{Zn}^{2+} + 2\mathrm{e}^{-} \rightarrow \mathrm{Zn}$
