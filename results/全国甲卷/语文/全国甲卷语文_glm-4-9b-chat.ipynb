【小题2】根据原文的句子“其基本做法为：首先将“假柱”（即临时的竖向支撑构件）安装在梁底部、原柱（原有立柱）旁边；再抽去原柱，使梁传来的重量暂时由“假柱”承担；然后安装新柱，新柱的材料、尺寸及安装位置与原有立柱相同；最后将“假柱”移去。”因此“假柱”被安装在原柱旁边。故填原柱。“加固过程中，技术人员虽然将原柱抽去，但是预先将“假柱”设置于原柱附近，让“假柱”代替原柱发挥支撑作用，因而换柱过程对结构整体的稳定基本无影响。换柱完成后，新柱与原柱有着同样的材料、尺寸，且与梁有着相同的可靠连接方式，它完全能够代替原柱发挥支撑作用。”既新柱可以代替原柱，故填新柱。

其次，坦诚交流有利于促进友谊。朋友之间，最需要的就是真诚相待。只有通过坦诚的交流，才能建立起深厚的友谊。在交往过程中，我们要勇于面对自己的不足，敢于承认错误，这样才能赢得他人的信任。同时，我们要倾听朋友的心声，给予关心和支持，这样才能让友谊历久弥新。

再次，坦诚交流有助于个人成长。在交流过程中，我们不仅要表达自己的想法，还要倾听他人的意见。通过对比、反思，我们可以发现自己的不足，从而不断完善自己。同时，在坦诚交流中，我们还能学到很多新知识，拓宽自己的视野。总之，坦诚交流是个人成长的重要途径。

再次，我们要学会倾听。倾听是交流的基础，只有倾听他人，才能真正了解他们的想法。在交流中，我们要用心倾听，不要急于发表意见，以免打断对方的思路。

总之，坦诚交流是人际交往中不可或缺的品质。只有通过坦诚的交流，我们才能迎来真正的相遇，让友谊地久天长。让我们勇敢地打开心灵之窗，用真诚和坦率书写美好的人生篇章。


```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "THUDM/glm-4-9b-chat"

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    torch_dtype=torch.float16,
).eval().to(device)
```

```python
subject, paper_type = "语文", "全国甲卷"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        if data["id"] == "98d815cb26fff705cd170a11":
            gen_kwargs = {"max_length": 2048, "do_sample": True, "temperature": 1.0, "top_p": 0.9}
        else:
            gen_kwargs = {"max_length": 2048, "do_sample": False}
        
        inputs = tokenizer.apply_chat_template(
            [{"role": "user", "content": question}],
            add_generation_prompt=True,
            tokenize=True,
            return_tensors="pt",
            return_dict=True
        )
        inputs = inputs.to(device)

        with torch.no_grad():
            outputs = model.generate(**inputs, **gen_kwargs)
            outputs = outputs[:, inputs['input_ids'].shape[1]:]
            
            response = tokenizer.decode(outputs[0], skip_special_tokens=True)

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("GLM4 9B 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
