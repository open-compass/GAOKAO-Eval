## 题目标答
【小题1】考察学生对于材料阅读内容的理解和分析。在原文中提到“在全球层面，构建海洋命运共同体主要着眼于涉海全球公共问题的治理，各国应致力于构建更加公正合理、稳定有效的国际法和国际规则体系。”实际上，原文并没有明确指出全球层面的国际法和国际规则体系不够完备或效率不高，也没有提到缺乏相应的合作制度体系。“总的来说，全球层面的海洋命运共同体构建，应着眼于建立更加有效的合作制度体系。”相反，文章强调了各国需要遵守国际法和国际规则，加强合作，共同应对海洋资源利用和管理所面临的挑战。

本题中，“刘备为吕布所袭奔于操程昱以备有雄才劝操图之”，这句话中，“刘备为吕布所袭”，主语是“刘备”，动词谓语是“为……所袭”，宾语是“于操”，那么，应在“刘备”和“于操”后断开，这就确定了两个断点：B、C。再看后面，“程昱以备有雄才”，意思是“程昱认为刘备有雄才”，那么，主语就应该是“程昱”，应在“程昱”后确定一个断点：E。这就已经确定了三个断点。再看“以备有雄才劝操图之”，意思是“（程昱）用（刘备）有雄才这一点劝曹操谋取（他）”，其中，主语是“程昱”，可用（程昱）来代换，动词谓语是“劝”，宾语是“之”，那么，应在“（程昱）”和“之”后确定断点，这就又确定了两个断点：F、H。

乘坦诚交流之翼，赴星辰大海之约

山和山不相遇，人和人要相处。人是社会关系的总和，而在处理纷繁复杂的相处时，有些人为避免冲突而不愿表达自己的想法。如果不能坦诚交流，或许能偶尔缓解尴尬，也确实能暂时与人相安无事，但这样只是带着面具的社交，没有温度也没有深度。而只有展开坦诚交流的羽翼，我们才能跨过繁文缛节的沟壑，相遇于星辰大海，实现心灵的双向奔赴。

首先，坦诚交流能够减少误解和误判。在人际交往中，如果我们总是隐藏自己的真实想法，他人就无法准确理解我们的意图。这样的结果往往是双方都在猜测对方的心思，导致误解的发生。而通过坦诚的交流，我们可以直接表达自己的观点和需求，他人也能从中获取准确的信息，从而减少误解的可能性。例如，在工作团队中，成员如果能够坦诚地表达自己的工作期望和担忧，团队合作将更加顺畅，工作效率也会得到提升。

```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "01-ai/Yi-1.5-34B-Chat"

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    torch_dtype=torch.float16,
).eval().to(device)
```

```python
subject, paper_type = "语文", "全国甲卷"
file_name = f"../data/{paper_type}/{subject}.jsonl"

terminators = [
    tokenizer.eos_token_id,
    tokenizer.convert_tokens_to_ids("<|eot_id|>"),
    tokenizer.convert_tokens_to_ids("<|im_end|>")
]

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        if data["id"] == "98d815cb26fff705cd170a11":
            gen_kwargs = {"max_length": 2048, "do_sample": True, "temperature": 1.0, "top_p": 0.9}
        else:
            gen_kwargs = {"max_length": 2048, "do_sample": False}

        input_ids = tokenizer.apply_chat_template(conversation=[{"role": "user", "content": question}], 
                                                tokenize=True, 
                                                add_generation_prompt=True, 
                                                return_tensors='pt')
        output_ids = model.generate(input_ids.to(device), eos_token_id=terminators, **gen_kwargs)
        response = tokenizer.decode(output_ids[0][input_ids.shape[1]:], skip_special_tokens=True)

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("Yi 34B 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
