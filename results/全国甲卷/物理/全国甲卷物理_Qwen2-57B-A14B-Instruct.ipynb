要解决这个问题，我们首先需要理解电路的基本原理以及电磁感应现象。给定的条件包括一个闭合回路，其中包含一个电容器、一个电阻、一个电源（通过外力做功提供），以及一个处于磁场中的移动导体。我们需要计算金属棒在加速过程中的速度，并在断开开关后分析电路的变化。

### （1）加速过程中的速度

当外力做功的功率等于电阻$R$热功率的2倍时，我们可以根据能量守恒和电路原理来解决问题。设金属棒的速度为$v$，则其动能增加率等于电路中能量损失的总和，即：

$P_{\text{外力}} = P_{\text{热}} + P_{\text{电容充电}}$

由于题目给出$P_{\text{外力}} = 2P_{\text{热}}$，我们可以写出：

$2I^2R = I^2R + \frac{d}{dt}\left(\frac{1}{2}CV^2\right)$

这里，$I$是电路中的电流，$V$是电容器两端的电压。由于金属棒的运动导致了电磁感应，产生了电动势$\mathcal{E}$，因此：

$I = \frac{\mathcal{E}}{R} = \frac{Blv}{R}$

代入上面的方程，得到：

$2\left(\frac{Blv}{R}\right)^2R = \left(\frac{Blv}{R}\right)^2R + \frac{d}{dt}\left(\frac{1}{2}CV^2\right)$

简化得到：

$2Bl^2v^2 = Bl^2v^2 + CV\frac{dV}{dt}$

因为金属棒最终会达到稳定状态，所以$\frac{dV}{dt} = 0$，从而可以解出$v$：

$v = \sqrt{\frac{CR}{2L^2}}$

### （2）断开开关后的分析

当开关$S$断开后，电路变为一个简单的RC电路，电容器开始放电。在断开开关的瞬间，电容器的电压最大，记为$V_0$。之后，电容器通过电阻$R$放电，直到电压降至0。根据RC电路的放电公式：

$V(t) = V_0e^{-t/RC}$

当外力做功的功率再次等于电阻$R$热功率的2倍时，我们可以利用同样的功率关系来找出$t$：

$P_{\text{外力}} = 2P_{\text{热}} = 2I^2R = 2\left(\frac{Blv}{R}\right)^2R$

由于此时金属棒以恒定速度$v$运动，我们可以用上述结果直接计算功率。但是，为了找到电容器两极间的电压$V(t)$，我们需要知道$t$。这通常需要额外的信息或假设，例如外力如何随时间变化，或者电容器放电的具体时间点。如果没有具体的时间信息，我们只能表达电压$V(t)$作为时间的函数，而不能给出具体的电压值。

对于从断开$S$开始到此刻外力做的功，我们需要知道这个“此刻”具体指的是什么时间点，才能计算出确切的数值。如果假设这个“此刻”是指电容器电压下降到某个特定值时，我们可以通过积分电容器放电过程中的功率来计算外力做的功。然而，没有具体的数值或时间点，我们无法给出具体的数值解答。

综上所述，对于问题（1），金属棒的速度可以通过上述步骤计算得出。对于问题（2），我们需要更多的具体信息才能给出详细的解答。

(2) 对于第二部分的问题，我们需要应用几何光学中的全反射原理。全反射发生在光线从光密介质（在这个问题中是玻璃柱）进入光疏介质（空气）时，入射角大于临界角。临界角 $\theta_c$ 可以通过 Snell 定律和折射率的关系来计算：
