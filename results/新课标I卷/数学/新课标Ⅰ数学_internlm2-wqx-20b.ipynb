因为 $  \frac { z } { z - 1 } = 1 + \text{i} $  ， 所以$ z=(z-1)(1+\text{i} )$ ，即$ z\cdot \text{i} =1+\text{i} $ ， 故$  z = \frac { 1 +\text{i} } {\text{i} } = \frac { ( 1 +\text{i} ) \cdot ( -\text{i} ) } {\text{i} \cdot ( -\text{i} ) } = 1 -\text{i} $ ． 故选$ \text{C}$ ．
## InternLM-WQX 模型输出
给定复数方程 $ \frac { z } { z - 1 } = 1 + \text{i}$，我们的目标是解出 $z$。

首先，我们可以将方程两边都乘以 $z - 1$ 来消除分母，得到：
$ z = (1 + \text{i})(z - 1) $

接下来，我们展开右边的乘积：
$ z = z - 1 + \text{i}z - \text{i} $

现在，我们将所有含 $z$ 的项移到方程的一边，常数项移到另一边：
$ z - \text{i}z = -1 + \text{i} $

合并同类项，得到：
$ (1 - \text{i})z = -1 + \text{i} $

为了求解 $z$，我们需要除以复数 $1 - \text{i}$ 的共轭复数，即 $1 + \text{i}$，以保持复数的实部：
$ z = \frac{-1 + \text{i}}{1 - \text{i}} $

为了除以复数，我们需要乘以共轭复数的倒数，即：
$ z = \frac{-1 + \text{i}}{1 - \text{i}} \cdot \frac{1 + \text{i}}{1 + \text{i}} $

展开分子和分母：
$ z = \frac{(-1 + \text{i})(1 + \text{i})}{(1 - \text{i})(1 + \text{i})} $

计算分子和分母：
分子：$(-1 + \text{i})(1 + \text{i}) = -1 + \text{i} - \text{i} - 1 = -2$
分母：$(1 - \text{i})(1 + \text{i}) = 1 - \text{i}^2 = 1 + 1 = 2$

所以，$z$ 等于：
$ z = \frac{-2}{2} = -1 $

但是，我们还需要考虑虚部。由于我们乘以了 $1 + \text{i}$，虚部会增加一倍。所以，$z$ 的虚部是 $1$ 的两倍，即 $2$。因此，$z$ 是 $-1 + 2\text{i}$，简化后为 $-1 + \text{i}$。

答案是：B. $-1+\text{i}$

$  \overrightarrow { b } - 4 \overrightarrow { a } = ( 2 , x - 4 )$ ，$  \overrightarrow { b } \bot \left( \overrightarrow { b } - 4 \overrightarrow { a } \right)=0$ ， ∴$ \overrightarrow { b } \left( \overrightarrow { b } - 4 \overrightarrow { a } \right) = 0$ ， ∴$ 4+x(x-4)=0$ ， ∴$ x=2$ ． 故选$ \text{D}$ ．
## InternLM-WQX 模型输出
已知向量 $\overrightarrow{a}=(0,1)$ 和 $\overrightarrow{b}=(2,x)$，我们要找的是使得 $\overrightarrow{b}$ 垂直于向量 $\overrightarrow{b}-4\overrightarrow{a}$ 的 $x$ 的值。

首先，我们计算向量 $\overrightarrow{b}-4\overrightarrow{a}$：
$\overrightarrow{b}-4\overrightarrow{a} = (2,x) - 4(0,1) = (2,x) - (0,4) = (2, x-4)$

两个向量垂直意味着它们的点积为0。因此，我们有：
$\overrightarrow{b} \cdot (\overrightarrow{b}-4\overrightarrow{a}) = 0$
$\Rightarrow (2,x) \cdot (2, x-4) = 0$
$\Rightarrow 2 \cdot 2 + x \cdot (x-4) = 0$
$\Rightarrow 4 + x^2 - 4x = 0$
$\Rightarrow x^2 - 4x + 4 = 0$

$(x-2)^2 = 0$

因此，$x-2=0$，解得 $x=2$。

所以，正确答案是 D. $2$。

$\cos(\alpha +\beta ) = \cos\alpha \cos\beta - \sin\alpha \sin\beta = m$

接着，我们使用两角差的余弦公式来表示 $\cos(\alpha -\beta )$：
$\cos(\alpha -\beta ) = \cos\alpha \cos\beta + \sin\alpha \sin\beta$

现在，我们想找到 $\cos\alpha \cos\beta$ 和 $\sin\alpha \sin\beta$ 的关系。由于 $\tan\alpha \tan\beta = 2$，我们可以写出：
$\frac{\sin\alpha}{\cos\alpha} \cdot \frac{\sin\beta}{\cos\beta} = 2$

$\sin\alpha \sin\beta = 2\cos\alpha \cos\beta$

$m = \cos\alpha \cos\beta - 2\cos\alpha \cos\beta$

$\sin\alpha \sin\beta = 2\cos\alpha \cos\beta$

$\cos(\alpha -\beta ) = \cos\alpha \cos\beta + \sin\alpha \sin\beta$
$\cos(\alpha -\beta ) = -\frac{m}{3} - \frac{2m}{3}$

```python
import re
import json

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = torch.device("cuda")

model_path = "internlm/internlm2-wqx-20b"
gen_kwargs = {"max_length": 2048, "do_sample": False}

tokenizer = AutoTokenizer.from_pretrained(
    model_path, trust_remote_code=True,
)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    torch_dtype=torch.float16,
).eval().to(device)
```

```python
subject, paper_type = "数学", "新课标卷Ⅰ"
file_name = f"../data/{paper_type}/{subject}.jsonl"

questions = []

print(f"试卷名：{paper_type} 高考真题 【{subject}】学科")

with open(file_name, "r") as f:
    for i, line in enumerate(f):
        data = json.loads(line)
        has_img, question = False, data['prompt']

        if '<img' in question:
            has_img = True
            question = re.sub(r'<img[^>]*?/>', "", question)

        inputs = tokenizer(question, return_tensors="pt")
        input_ids = inputs["input_ids"].to(device)

        outputs = model.generate(input_ids, **gen_kwargs)
        outputs = outputs[0].cpu().tolist()[len(inputs["input_ids"][0]) :]

        response = tokenizer.decode(outputs, skip_special_tokens=True)
        response = response.split("<|im_end|>")[0]

        if i == 0:
            print("*" * 35)
        else:
            print("*" * 15)

        print("题目编号：" + str(i+1) + ("（含图片）" if has_img else ""))
        print("题目标答：" + data["answer"])
        print("InternLM-WQX 模型输出：" + response)
            
        questions.append({
            "id": str(i+1),
            "question": question,
            "answer": data["answer"],
            "output": response,
            "has_img": has_img
        })
```
